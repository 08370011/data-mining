```python
dfnew=df[['forumName','content','gender']]
```

```python
dfnew['content']=dfnew['content'].astype(str)
def cutword(x):
    seg=jieba.cut(x,cut_all=False)
    return ','.join(seg)
dfnew['content']=dfnew['content'].apply(cutword)
```

```python
dfnew
```

```python
dfnew.groupby(['forumName']).size()
```

```python
merge=dfnew.groupby("forumName")['content'].apply(''.join).reset_index()
```

```python
merge
```

```python
corpus = merge['content'].values.tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer  
vectorizer = CountVectorizer()  
X = vectorizer.fit_transform(corpus)  
word = vectorizer.get_feature_names()  

from sklearn.feature_extraction.text import TfidfTransformer  
transformer = TfidfTransformer()  
tfidf = transformer.fit_transform(X)
```

```python
print(tfidf )
```

```python
len(corpus)
features = tfidf.toarray()
labels = merge['forumName'].values
labels
```

```python
textFrame = pd.DataFrame(features, columns = word)
textFrame.head()
len(features)
```

```python
m = 0.01
dataset = list()
for i in range(len(features)):
    temp = textFrame.columns.values[[textFrame.iloc[i] >= m]]
    dataset.append(temp.tolist())
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
graph = df.T
graph.columns = merge['forumName']
```

```python
graph.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

result=apriori(graph, min_support = 0.01, use_colnames=True)
```

```python
rules = association_rules(result, metric="confidence", min_threshold=0.8)
rules.head()
```
