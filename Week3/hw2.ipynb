```python
dfnew=df[['forumName','content','gender']]
```

```python
dfnew['content']=dfnew['content'].astype(str)
def cutword(x):
    seg=jieba.cut(x,cut_all=False)
    return ','.join(seg)
dfnew['content']=dfnew['content'].apply(cutword)
```

```python
dfnew
```

```python
merge=dfnew.groupby("forumName")['content'].apply(''.join).reset_index()
```

```python
list1=jieba.cut(merge['content'],cut_all=False)
```

```python
for i in list1:
    print(i)
```

```python
corpus = merge['content'].values.tolist()
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer  
vectorizer = CountVectorizer()  
X = vectorizer.fit_transform(corpus)  
word = vectorizer.get_feature_names()  

from sklearn.feature_extraction.text import TfidfTransformer  
transformer = TfidfTransformer()  
tfidf = transformer.fit_transform(X)
```

```python
len(corpus)
features = tfidf.toarray()
labels = merge['forumName'].values
labels
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
transData = pca.fit_transform(features, labels)
type(transData)
merge['pca1'] = DataFrame(transData[:,0])
merge['pca2'] = DataFrame(transData[:,1])
merge.head()
```
